#  텍스트 생성

## 문장 생성

```python
# 프롬프트 준비
# prompt = '''다음 이야기를 써주세요.
# 기타를 좋아하지만 컴맹인 여고생이 어떤 계기로 록밴드에 가입하고, 낯선 인간관계를 통해 활동하게 되는 이야기.'''

# 프롬프트 준비
prompt = '''다음 이야기를 써주세요.
밴드부에 도전하는 롤 프로게이머 스토리'''
```

```python
import openai

# 텍스트 생성 실행
response = openai.Completion.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prompt,
    temperature=0.7,
    max_tokens=500
)
print(response["choices"][0]["text"])
```

```python
print(response)
```

# 채팅

```python
# 채팅 메시지 리스트 준비
messages = [
    {"role": "system", "content": "롤 프로게이머이고 성격이 매우 소심함. 게임 용어를 섞어서 사용하기도 하고 음침한 성격의 소유자, 누가 말걸면 불쾌해함"},
    {"role": "user", "content": "너가 좋아하는 게임이 뭐니?"},
]
```

```python
import openai

# 채팅 실행
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages=messages,
    temperature=1.2
)
response["choices"][0]["message"]["content"]
```

```python
print(response)
```

# 삽입

```python
# 프롬프트 준비
prefix_prompt = """def helloworld():
    '''
    설명: """

suffix_prompt = """
    '''
    print("Hello World!")

helloworld()
"""
```

```python
import openai

# 삽입 실행
response = openai.Completion.create(
    model="gpt-3.5-turbo-instruct",
    prompt=prefix_prompt,
    suffix=suffix_prompt,
    temperature=0.7,
    max_tokens=300
)
print(response["choices"][0]["text"])
```

# 편집

```python
# 채팅 메시지 리스트 준비
messages = [
    {"role": "system", "content": "오타를 수정해 주세요."},
    {"role": "user", "content": "오늘은 정말 즐거웠따."},
]
```
