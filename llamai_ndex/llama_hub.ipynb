```python
from llama_index.core import VectorStoreIndex, download_loader

from llama_index.readers.web import BeautifulSoupWebReader

loader = BeautifulSoupWebReader()
documents = loader.load_data(urls=["https://news.naver.com/"])
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine(streaming=True)
streaming_response = query_engine.query("이 웹페이지에서 전하고 싶은 말은 무엇인가요? 한국어로 대답해 주세요.")
streaming_response.print_response_stream()
# index.query("What language is on this website?")
```

```python
# 쿼리 엔진 생성
query_engine = index.as_query_engine()
```

```python
# 질의응답
print(query_engine.query("이 웹페이지에서 전하고 싶은 말은 무엇인가요? 한국어로 대답해 주세요."))
```

```python
from llama_index.readers.youtube_transcript import YoutubeTranscriptReader

loader = YoutubeTranscriptReader()
documents = loader.load_data(
    ytlinks=["https://www.youtube.com/watch?v=oc6RV5c1yd0"]
)
```

```python
# from llama_index.readers.youtube_transcript.utils import is_youtube_video

# is_youtube_video("https://youtube.com/watch?v=j83jrh2")  # => True
# is_youtube_video("https://vimeo.com/272134160")  # => False
```

```python
index = VectorStoreIndex.from_documents(documents)
query_engine = index.as_query_engine(streaming=True)
streaming_response = query_engine.query("이 동영상에서 전하고 싶은 말은 무엇인가요? 한국어로 대답해 주세요.")
streaming_response.print_response_stream()
# print(query_engine.query("이 동영상에서 전하고 싶은 말은 무엇인가요? 한국어로 대답해 주세요."))
```
